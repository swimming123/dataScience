### 2. 아래 요구사항에 맞게 코드를 작성해주세요.
- 아래의 예시와 같이 name, age의 변수를 포함한 Person 클래스를 만들어주세요.(2점)

```
 # 여기에 코드를 작성해주세요.
class Person:
     name = ""
     age = 0

     def __init__(self, name, age):
         self.name = name
         self.age = age
 
```

```
person = Person('Liam', '21')
```

```
person.name, person.age
```

- 위의 Person클래스를 상속받아 play의 메소드를 추가한 Child 클래스를 만들어주세요(3점)

  play: print('{} is playing.'.format(name))

```
 # 여기에 코드를 작성해주세요.
class Child (Person):
  def play(self):
    print("{} is playing..".format(self.name))
```

```
# 아래 코드는 실행 후 결과 예 입니다.
child = Child('Olivia', '8')
```
