# [6주차 코딩 실습](04. MySQL_45. 테이블 객체 생성을 위한 클래스 작성 및 저장)강의자료를 참고하여 아래 조건에 맞게 컬럼데이터를 작성하고, 테이블 객체 생성을 위한 클래스를 생성해보세요.(20점)

* 클래스명 : EPL, 테이블 이름 : EPL1

* index : column(Integer, primary_key = True)

* name : column(string(20))

* gf = column(Integer)

* ga = column(Integer)





```
from sqlalchemy import *
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
```

```
# 여기에 코드를 작성하세요
class EPL():
  __tablename__ = "EPL1"

  #컬럼 데이터 작성 
  index = column(Integer)
  name = column(String(20))
  gf = column(Integer)
  ga = column(Integer)

  #생성자 함수
  def __init__(self, index, name, gf, ga):
    self.index = index
    self.name = name
    self.gf = gf
    self.ga = ga

  #repr 함수
  def __repr__(self):
    return "<User {}, {}, {}, {}>".format(self.index, self.name, self.gf, self.ga)
```
